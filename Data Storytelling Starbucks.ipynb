```python
# https://stackoverflow.com/questions/60498999/line2d-object-has-no-property-kind

fig,ax = plt.subplots(1)
ax = df_sugars.groupby(['Beverage_category']).mean().plot(kind='barh')
#ax.set_xlabel('x label name')   # replace with the labels you want
ax.set_xlabel('Mean Sugar Content (g)')
#plt.xticks(rotation=30)
plt.axvline(x = 25, color = 'r', linestyle = '-', label = 'women')
plt.axvline(x = 36, color = 'g', linestyle = '-', label = 'men')
ax.legend()
plt.show()
fig.savefig("Starbucks Drinks Sugar Content.png")
```

```python
# From above, we can see that the average sugar content in 7 of the 9 drink categories are above the total amount
# of sugar a woman should consume in a day, and the average content in 4 of the 9 drink categories are above
# the amount a man should consume in the course of a day !  These are all assumed to be added sugars without 
# nutritional benefit.
# Guideilnes here https://www.heart.org/en/healthy-living/healthy-eating/eat-smart/sugar/how-much-sugar-is-too-much
```
